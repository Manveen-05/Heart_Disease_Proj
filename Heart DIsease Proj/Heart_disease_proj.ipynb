#Model Evaluation/Checking


```python
# Import function to split dataset into training and testing sets
from sklearn.model_selection import train_test_split

# Import StandardScaler for feature scaling (normalizing data)
from sklearn.preprocessing import StandardScaler

# Import evaluation metrics to measure model performance
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Import Logistic Regression model (linear model for classification)
from sklearn.linear_model import LogisticRegression

# Import Gaussian Naive Bayes model (probabilistic classifier)
from sklearn.naive_bayes import GaussianNB

# Import Decision Tree Classifier (tree-based model)
from sklearn.tree import DecisionTreeClassifier

# Import Support Vector Classifier (SVM model)
from sklearn.svm import SVC

# Import K-Nearest Neighbors classifier (distance-based model)
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Drop target column
X = data_encode.drop('HeartDisease', axis=1)

# Target variable
y = data_encode['HeartDisease']
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,      # 20% data for testing
    random_state=42     # for reproducibility
)
```

```python

scaler = StandardScaler()

# Fit ONLY on training data, then transform
X_train = scaler.fit_transform(X_train)

# Use same scaler for test data
X_test = scaler.transform(X_test)
```

```python
# Dictionary of models
models = {
    "Logistic Regression": LogisticRegression(),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "Decision Tree": DecisionTreeClassifier(),
    "SVM": SVC(),
}
```

```python
results = []
```

```python
results.clear() # Clear results from previous runs
for name, model in models.items():

    # Train model
    model.fit(X_train, y_train)

    # Predict on test data
    y_pred = model.predict(X_test)

    # Evaluate
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)

    # Store results
    results.append((name, acc, f1))
```

```python
print("\n--- Model Performance ---")
for name, acc, f1 in results:
    print(f"Model: {name}")
    print(f"Accuracy: {acc:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print("-"*30)
```

#Deploying


```python
# Train Final KNN Model
import joblib
# Access the trained KNN model from the 'models' dictionary
knn_model = models['KNN']
joblib.dump(knn_model, 'KNN_heart.pkl')
joblib.dump(scaler, 'scaler.pkl')
# Save column names to a different file to avoid overwriting the model
joblib.dump(X.columns.tolist(), 'features.pkl')
```
